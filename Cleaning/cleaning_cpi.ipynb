```python
# Dependencies
import pandas as pd
```

```python
# Path to CPI csv file, load in and read CPI data
cpi_file = '../Resources/USCPI_allitems.csv'

cpi_df = pd.read_csv(cpi_file)
cpi_df
```

```python
# Create a column detailing the mean CPI on a national level per year
cpi_df['Average CPI'] = cpi_df[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].mean(axis=1)
cpi_df = cpi_df[['Year', 'Average CPI']]
cpi_df
```

```python
# Load in and read clean HPI csv
clean_hpi_path = '../Resources/clean_monthly_hpi.csv'

hpi_df = pd.read_csv(clean_hpi_path)
hpi_df
```

```python
# Group HPI data by year and get the mean HPI... preparing to merge CPI and HPI by year
year_hpi_df = hpi_df.groupby(['Place Name', 'Year'], as_index= False)['Index (SA)'].mean()
us_year_hpi = year_hpi_df.loc[year_hpi_df['Place Name'] == 'United States', :]
us_year_hpi
```

```python
# Merge CPI and HPI data
us_cpi_hpi = pd.merge(us_year_hpi, cpi_df, on='Year')
us_cpi_hpi
```

```python
# Push the remade DataFrame to a new CSV file
us_cpi_hpi.to_csv('../Resources/US_CpiHpiAvg.csv', encoding='utf-8', index= False)
```
