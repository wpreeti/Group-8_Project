```python
# import dependancies 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# Path to CPI csv file, load in and read CPI data 
hpi_file = r"C:\Users\Sam Song\Desktop\Data Analytics Bootcamp\Project 1\Group-8_Project\Resources\Master_HPI_data.csv"


hpi_df = pd.read_csv(hpi_file)
hpi_df
```

```python
# Rename column headers
renamed_hpi_df = hpi_df.rename(columns={'hpi_type': 'HPI Type', 
                                        'hpi_flavor': 'HPI Flavor', 
                                        'frequency': 'Frequency', 
                                        'level': 'Level', 
                                        'place_name': 'Place Name',
                                        'place_id': 'Place ID', 
                                        'yr': 'Year', 
                                        'period': 'Period', 
                                        'index_nsa': 'Index (NSA)', 
                                        'index_sa': 'Index (SA)'})

renamed_hpi_df
```

```python
# Display only data for the U.S. and Census Divisions (monthly data)
monthly_hpi = renamed_hpi_df.loc[renamed_hpi_df['Frequency'] == 'monthly', :]
monthly_hpi
```

```python
# Exlcude HPI Type, Flavor, Frequency, Level of Geography, Place ID; All unneccessary for analysis
# Include only seasonally adjusted data
clean_monthly_df = monthly_hpi[['Place Name', 'Year', 'Period', 'Index (SA)']]
clean_monthly_df
```

```python
clean_monthly_df['Index (SA) % Change'] = clean_monthly_df['Index (SA)'].pct_change()
clean_monthly_df

```

```python
clean_monthly_df = clean_monthly_df.loc[clean_monthly_df["Period"] == 1]
clean_monthly_df
```

```python
# checking for all unique regions
(clean_monthly_df["Place Name"].unique())
```

```python
final_df = clean_monthly_df
final_df 
```

```python
# Push the remade DataFrame to a new CSV file
final_df.to_csv('../Group-8_Project/Resources/sam_final_df', index= False, encoding='utf-8')
```
